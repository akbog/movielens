```python
import pandas as pd
from datetime import datetime
```

```python
box_office = pd.read_csv("Box_Office_Data_1989_2018.csv")
box_office["gross"] = box_office.gross.astype('float')
box_office  = box_office.drop(9253, axis = 0)
box_office.release = box_office.apply(lambda x : datetime.strptime("{} {}".format(x.release, x.year), "%b %d %Y"), axis = 1)
```

```python
box_office
```

```python
#Dividing by Movie Seasons
#Winter (Dec 21st - March 21st)
#Spring (March 21st - June 21st)
#Summer (June 21st - September 21st)
#Fall (September 21st - Dec 21st)
for index, row in box_office.iterrows():
    year = row["year"]
    seasons = {"Q1" : (datetime.strptime("{} 01 01".format(year), "%Y %m %d"), (datetime.strptime("{} 03 31".format(year), "%Y %m %d"))),
               "Q2" : (datetime.strptime("{} 04 01".format(year), "%Y %m %d"), (datetime.strptime("{} 06 30".format(year), "%Y %m %d"))),
                "Q3" : (datetime.strptime("{} 07 01".format(year), "%Y %m %d"), (datetime.strptime("{} 09 30".format(year), "%Y %m %d"))),
                "Q4" : (datetime.strptime("{} 10 01".format(year), "%Y %m %d"), (datetime.strptime("{} 12 31".format(year), "%Y %m %d")))}
    if seasons["Q1"][0] <= row["release"] <= seasons["Q1"][1]:
        box_office.loc[index, "season"] = "Q1 " #+ str(year)
        box_office.loc[index, "time"] = (datetime.strptime("{} 03 31".format(year), "%Y %m %d"))
    elif seasons["Q2"][0] <= row["release"] <= seasons["Q2"][1]:
        box_office.loc[index, "season"] = "Q2 " #+ str(year)
        box_office.loc[index, "time"] = (datetime.strptime("{} 06 30".format(year), "%Y %m %d"))
    elif seasons["Q3"][0] <= row["release"] <= seasons["Q3"][1]:
        box_office.loc[index, "season"] = "Q3 " #+ str(year)
        box_office.loc[index, "time"] = (datetime.strptime("{} 09 30".format(year), "%Y %m %d"))
    elif seasons["Q4"][0] <= row["release"] <= seasons["Q4"][1]:
        box_office.loc[index, "season"] = "Q4 " #+ str(year)
        box_office.loc[index, "time"] = (datetime.strptime("{} 12 31".format(year), "%Y %m %d"))
```

```python
#Now we have the release schedule (can group by)
# box_office.season = box_office.apply(lambda x : "{} {}".format(x.season, str(x.year)), axis = 1)
```

```python
box_office.season.unique() #we want to drop all NAN
remove = ["nan {}".format(year) for year in range(1989, 2019)]
box_office = box_office.drop(["year"], axis = 1)
cleaned = box_office[~box_office['season'].isin(remove)]
cleaned = cleaned[~cleaned['studio'].isin(["-"])]
cleaned = cleaned.drop( ['time','studio'], axis = 1)
```

```python
cleaned
```

```python
#Now we want to Group by Season, then Studio, and sum Gross 
output = cleaned.groupby(["season", "studio"]).sum()
```

```python
output
```

```python
cleaned_export = cleaned[~cleaned['rank'].isin(range(21,cleaned.shape[0]))]
cols = cleaned_export.columns.tolist()
cols = cols[1:2] + cols[0:1] + cols[2:]
cleaned_exports = cleaned_export[cols]
cleaned_exports.to_csv("box_office_data_cleaned_2.csv", index = False)

# output.to_csv("seasonal_studio_revenue.csv")
```
